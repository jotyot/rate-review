```python
import pandas as pd

reviews = pd.read_json('data/ys-reviews-with-categories.json')
reviews = reviews.loc[reviews['category'] == "restaurant"]
reviews.head()
```

```python
from string import punctuation

keyword_groups = {
    'food': ['food', 'taste', 'sushi', 'pizza', 'mexico', 'ramen', 'buffet', 'fish', 
             'delicious', 'taco', 'bagels', 'fries', 'mexican', 'ice cream', 'chicken', 
             'dessert', 'burger', 'fresh', 'breakfast', 'salad', 'coffee', 'cheese', 'meal',
             'sandwiches', 'lasagna', 'margherita', 'wings', 'popcorn', 'sandwich', 'pie', 
             'yummy', 'steak', 'burritos', 'eggs', 'biscuit', 'sauce', 'bbq', 'meats', 'noodles',
             'rolls', 'pho', 'chinese', 'drinks', 'poke', 'fudge', 'milk', 'ate', 'eat',
             'pineapple', 'cream', 'japanese', 'mole', 'roll', 'margarita', 'veggie', 'tasty', 
             'goodness', 'sub', 'turkey', 'dogs', 'dog', 'soup', 'thai', 'vietnamese', 'lunch',
             'dinner', 'bean', 'rice', 'burrito', 'drink', 'ribs', 'juices', 'panini', 'iced', 
             'yum', 'waffles', 'italian', 'potatoes', 'nachos', 'salsa', 'chips', 'carne',
             'rigatoni', 'beers', 'beer', 'yum', 'dish', 'crust', 'calzone', 'lemonade', 'quality',
             'nuggets', 'wraps', 'wrap', 'rib', 'ribeyes', 'ribeye', 'lobster', 'crab', 'bone', 
             'calamari', 'cooked', 'shrimp', 'salmon', 'spicy', 'bland', 'curry', 'donut', 'doughnut',
             'sorbet', 'product', 'puffs', 'flavors', 'flavor', 'mein', 'pork', 'grill', 'dim',
             'crepe', 'dimsum', 'slices', 'shakes', 'shake', 'latte', 'bipimbap', 'chowder', 'oysters',
             'alfredo', 'alfrado', 'menu', 'apricot', 'tea', 'garlic', 'spaghetti', 'chalupa',
             'churros', 'mocha', 'latte'],
    'service': ['service', 'staff', 'servers', 'person', 'server', 'waitress', 'waiter', 
                'minutes', 'baristas', 'order', 'people', 'friendly', 'fast', 'wait', 'slow',
                'rude', 'min', 'bartender', 'bartenders', 'long', 'man', 'barista', 'employees'],
    'place': ['location', 'place', 'atmosphere', 'vibe', 'interior', 'comfortable', 'chaotic', 'busy', 
              'crowded', 'ambience', 'quiet'],
    'clean': ['clean', 'dirty', 'filthy', 'flies', 'cockroach'],
    'price': ['cheap', 'price', 'expensive', 'value', 'worth', 'pricey', 'pricy', 'expenses']
}

def removed_punct(text: str) -> str:
    # make lower case and remove all punctuations
    text = text.lower().translate(str.maketrans('', '', punctuation))
    return text

# Function to assign groups to a review
def assign_groups(review) -> list[str]:
    groups = []
    for group, keywords in keyword_groups.items():
        if any(keyword in removed_punct(review) for keyword in keywords):
            groups.append(group)
    return groups

reviews['keyword'] = [assign_groups(review) for review in reviews.text]

with open('review-keywords.json', 'w') as outfile:
    outfile.write(reviews.to_json())

print(reviews[reviews['keyword'].apply(len) == 0].__len__())

# print("\n".join(reviews[reviews['keyword'].apply(len) == 0].text.tolist()))
```
