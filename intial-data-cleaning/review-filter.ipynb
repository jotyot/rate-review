```python
import json

def scan():
    gmap_ids = yolosac.gmap_id.to_list()
    with open('data/review-California.json', 'r') as infile, open('review-all.json', 'w') as temp:
        total = 1
        for line in infile:
            data = json.loads(line)
            total += 1
            if data.get('text') is not None and data.get('gmap_id') in gmap_ids:
                out = json.dumps({key: data[key] for key in ['text', 'rating', 'gmap_id']})
                temp.write(out + '\n')
            if total % 100000 == 0: print(str(round(total * 100 / 70000000, 4)) + '%')

scan()
```

```python
import json
import math

def append(start_percent = 0.0):
    gmap_ids = yolosac.gmap_id.to_list()
    with open('data/review-California.json', 'r') as infile, open('review-temp.json', 'w') as temp:
        total = 1
        for _ in range(math.floor(70000000 * start_percent / 100)):
            total += 1
            if total % 100000 == 0: print(str(round(total * 100 / 70000000, 4)) + '%')
            next(infile)
        for line in infile:
            data = json.loads(line)
            total += 1
            if data.get('text') is not None and data.get('gmap_id') in gmap_ids:
                out = json.dumps({key: data[key] for key in ['text', 'rating', 'gmap_id']})
                temp.write(out + '\n')
            if total % 100000 == 0: print(str(round(total * 100 / 70000000, 4)) + '%')
```

```python
def reformatJson():
    with open('data/yolosac-reviews.json', 'r') as infile, open('data/yolosac-review.json', 'w') as outfile:
        outfile.write('[\n')
        lines = infile.readlines()

        # adds comma to every line except last
        for i, line in enumerate(lines):
            if i < len(lines) - 1: line = line.rstrip() + ',\n'
            outfile.write(line)

        outfile.write(']')
        
reformatJson()
```

### Filter out reviews from uncategorized businesses

In `reduce_categories.ipynb`, we group most of the businesses into 8 general categories, and remove extra businesses. We only want to remove reviews from these extra businesses so that each review is assigned a general category.

```python
# Open the file and load the JSON array
with open('data/yolosac-reviews.json', 'r') as file:
    data = json.load(file)

remove = False
prev_id = ''
category = ''
# Reviews in file are ordered by business already. This should speed up category assignment.
for review in reversed(data):
    # if review has different id than previous
    if review['gmap_id'] != prev_id:
        prev_id = review['gmap_id']   
        try:
            # try to lookup category
            category = yolosac[yolosac['gmap_id'] == review['gmap_id']]['category'].iloc[0]
            review['category'] = category
            remove = False
        except IndexError:
            data.remove(review)
            remove = True
    else:
        if remove:
            data.remove(review)
        else:
            review['category'] = category

# Sanity check to make sure we deleted unwanted reviews
print(len(data))
# Write to new JSON file
with open('data/ys-reviews-with-categories.json', 'w') as file:
    json.dump(data, file)
    file.close()
```

Save categorized reviews to JSON file and compress for portability.

```python
import zipfile

# Specify the file paths
source_file = 'data/ys-reviews-with-categories.json'
zip_file = 'ys-reviews-with-categories.zip'

# Create a Zip file and add the source file to it
with zipfile.ZipFile(zip_file, 'w',zipfile.ZIP_DEFLATED, compresslevel=9) as zipf:
    zipf.write(source_file)
```
