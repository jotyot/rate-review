```python
import pandas as pd
yolosac = pd.read_csv('yolosac.csv')
yolosac
```

```python
import json

def scan_mil(mil = 0):
    gmap_ids = yolosac.gmap_id.to_list()
    with open('data/review-California.json', 'r') as infile, open('review-temp.json', 'w') as temp:
        success, total = 0, 0
        for i in range(mil * 1000000):
            next(infile)
            print(i)
        for line in infile:
            data = json.loads(line)
            if data.get('text') is not None and data.get('gmap_id') in gmap_ids:
                temp.write(line)
                success += 1
            total += 1
            print(success, total)
            if total == 1000000: break
    
    with open('review-temp.json', 'r') as temp, open('yolosac-reviews.json', 'a') as outfile:
        for line in temp:
            outfile.write(line)

# would need to run scan_mil(0 to 70) and not repeat one each time...
# scan_mil(０) ran
scan_mil(0)
```
